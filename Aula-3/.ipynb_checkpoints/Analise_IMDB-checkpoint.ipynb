```python
color_or_bw = imdb.query("color in ['Color', ' Black and White']")
len(color_or_bw)
```

```python
resumo = color_or_bw['color']=='Color'
resumo.sample(15)
```

```python
color_or_bw['color_0_ou_1'] = (color_or_bw['color']=="Color") * 1
color_or_bw.head()
```

```python
color_or_bw["color_0_ou_1"].value_counts()
```

```python
color_or_bw.groupby("color").mean()["gross"]
```

```python
sns.catplot(data=color_or_bw, x="color", y="gross")
plt.title("Faturamento dos filmes por cor")
plt.show()
```

## Veja que a média de notas dos filmes preto e branco é maior que de filmes coloridos
Provavelmente pq apenas cinéfilos e amantes do cinema, vão assistir estes filmes e avalia-los como obras-prima

```python
color_or_bw.groupby("color").mean()["imdb_score"]
```

```python
color_or_bw.groupby("color").median()["imdb_score"]
```

```python
color_or_bw.query("color_0_ou_1==1")['imdb_score'].plot(kind='hist')
plt.title('Distribuição de notas para filmes coloridos')
plt.show()
```

```python
color_or_bw.query("color_0_ou_1==0")['imdb_score'].plot(kind='hist')
plt.title('Distribuição de notas para filmes Preto e branco')
plt.show()
```

## Existe correlação entre orçamento e faturamento?

```python
orc_fat = imdb[['budget', 'gross']].dropna().query("budget>0 | gross>0")
sns.scatterplot(data=orc_fat, x="budget", y="gross")
plt.show()
```

> A escala esta dificultando a analise

Para melhorar esta analise, vamos analisar apenas filmes americanos

### Análise de Filmes americanos

```python
filmes_usa = imdb.query("country=='USA'")
filmes_usa.sort_values("budget", ascending=False).head()
```

### Remover Duplicados

```python
filmes_usa = filmes_usa.drop_duplicates()
```

```python
orc_fat = filmes_usa[['content_rating', 'budget', 'gross']].dropna().query("budget>0 | gross>0")
```

```python
sns.scatterplot(data=orc_fat, x="budget", y="gross")
plt.show()
```

# Prejuizo X Lucro

```python
filmes_usa['lucro'] = filmes_usa['gross'] - filmes_usa['budget'] # Add nova coluna "lucro"

budget_gross = filmes_usa.query("budget >0 | gross > 0")[["budget", "lucro"]].dropna()

sns.scatterplot(x="budget", y="lucro", data=budget_gross)
```

```python
budget_gross = filmes_usa.query("budget >0 | gross > 0")[["title_year", "lucro"]].dropna()

sns.scatterplot(x="title_year", y="lucro", data=budget_gross)
```

```python
filmes_por_diretor = filmes_usa['director_name'].value_counts()
filmes_por_diretor.head()
```

```python
gross_director = filmes_usa[['director_name', 'gross']].set_index("director_name").join(filmes_por_diretor, on="director_name")
gross_director.columns=["Grana", "Qtd_Filmes"]
gross_director = gross_director.reset_index()
gross_director.head()
```

```python
sns.scatterplot(data=gross_director, x="Qtd_Filmes", y="Grana")
```

```python
sns.pairplot(data= filmes_usa[["gross", "budget", "lucro", "title_year"]])
```

```python
filmes_usa[["gross", "budget", "lucro", "title_year"]].corr()
```

```python
sns.heatmap(filmes_usa[["gross", "budget", "lucro", "title_year"]].corr())
```

```python
graf = gross_director.dropna().query("Grana>0").sort_values("Qtd_Filmes", ascending=False).head(150)
graf.sample(5)
```

```python
sns.scatterplot(data=graf, x="Qtd_Filmes", y="Grana", hue="director_name", size="director_name", sizes=(30,300))
plt.gcf().set_size_inches(24,12)
plt.show()
```

```python
sns.relplot(data=graf, x="director_name", y="Grana", hue="Qtd_Filmes", size="Qtd_Filmes", sizes=(30,300))
plt.gcf().set_size_inches(20,10)
plt.show()
```

```python
orc_fat = filmes_usa[['content_rating', 'budget', 'gross']].dropna().query("budget>0 | gross>0")
sns.scatterplot(data=orc_fat, x="budget", y="gross", hue="content_rating", size="content_rating", sizes=(15,250))
plt.gcf().set_size_inches(20,10)
plt.show()
```

```python
orc_fat.groupby("content_rating").count().sort_values("budget", ascending=False)
```

> Existe grande predominancia dos filmes classificados como R e PG-13 na quantidade de filmes produzidos

```python
sns.catplot(data=orc_fat, x="content_rating", y="gross", kind="swarm")
plt.gcf().set_size_inches(20,10)
```
