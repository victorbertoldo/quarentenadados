```python
import seaborn as sns
```

```python
df_exemplo = sns.load_dataset("tips")
df_exemplo.head()
```

```python
sns.relplot(x="total_bill", y="tip", data=df_exemplo);
plt.show()
```

### A visualização acima, nos mostra  que há uma relação entre o valor da conta e o valor da gorjeta.
#### Quanto mais cara a conta, mais alta a gorjeta

## Trabalhando com variáveis de semantica. Usando o campo smoker para colorir de acordo com o campo.

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=df_exemplo)
plt.show()
```

### Agora Usando o campo sex para colorir de acordo com o campo.

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=df_exemplo)
```

### Incrementando com o parametro style

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=df_exemplo)
```

### Colorir o grafico por uma coluna quantitativa:

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=df_exemplo)
```

## Aumentando os pontos por variavel qualitativa

```python
sns.relplot(x="total_bill", y="tip",  hue="size", size="size", sizes=(15,150), data=df_exemplo)
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=df_exemplo)
```

### Redimensionando Figuras no Seaborn

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(15, 7))

sns.scatterplot(x="total_bill", y="tip", hue="day", data=df_exemplo)
```

### Usando cores sequenciais para plotar um grafico

```python
sns.relplot(x="total_bill", y="tip", hue="size", palette="Blues", size="size", sizes=(12,250), data=df_exemplo)
plt.show()
```

### Usando Cores qualitativas

```python
sns.catplot(x="sex", kind="count", palette="Set2", data=df_exemplo)
```

### Cores divergentes

```python
sns.heatmap(df_exemplo.corr(), annot=True, cmap="RdYlBu")
```

## Definindo a paleta padrão do Seaborn

```python
sns.set_palette("pastel")

sns.relplot(x="total_bill", y="tip", data=df_exemplo)
```

### Utilizando o with para definir a paleta temporariamente

```python
with  sns.color_palette("Accent"):
    sns.relplot(x="total_bill", y="tip", size="size", sizes=(12,250), data=df_exemplo)
```

## Gráficos de barras

```python
import pandas as pd

day_count = pd.crosstab(df_exemplo.day, df_exemplo.size, dropna=True)
pd.DataFrame(day_count)
day_count.columns = ["Total"]
day_count = day_count.reset_index().sort_values("Total", ascending=False)

# pegadinha do parametro order é q ele ordena as variaveis categoricas
sns.catplot(x="day", y="Total", kind="bar", palette="Set2", data=day_count, order=day_count["day"].values.tolist()) 
plt.show()
```

### Barras na horizontal

```python
sns.catplot(y="day", kind='count', palette="Set3", data=df_exemplo)
```

### Sem passar parametro kind=

```python
sns.catplot(x="day", y="total_bill", palette="Set1", data=df_exemplo)
plt.show()
```

### Melhorando a viiz acima, minimizando a sobreposição de pontos

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=df_exemplo)
plt.show()
```

### Adicionando o parametro Hue à viz

```python
sns.catplot(x="day", y="total_bill", kind="swarm", hue="sex", palette="Dark2", data=df_exemplo)
```

> Fica claro que os valores mais altos são geralmente pagos por homens

# Adicionando linhas de regressão para visualizar a dispersão entre variaveis

```python
%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="sex", palette="Pastel1", data=df_exemplo)
plt.show()
```

```python
%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", palette="Set2", data=df_exemplo)
```

```python
%matplotlib inline
sns.lmplot(x="total_bill",
           y="tip",
           hue="smoker",
           col="time",
           palette="Set1",
           data=df_exemplo,
           scatter_kws={'s': 90, 'alpha': 0.5},
           line_kws={'lw':4})
```

```python
%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", palette="Set2", data=df_exemplo)
```

## Visualizando Outliers

```python
sns.catplot(x="day", y="total_bill", kind="box",data=df_exemplo)
plt.show()
```

### Visualizar se a pessoa é fumante ou não

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=df_exemplo)
```

### Dividindo o grafico em 2 para analisar separadamente por sexo

```python
sns.catplot(x="day", y="total_bill", kind="box", col="sex", data=df_exemplo)
plt.show()
```

## Visualizando outro tipo de boxplot
este não dá uma visão tão boa dos outliers, mas se o volume de dados for muito grande, é o ideal para visulizar

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=df_exemplo)
```

```python
sns.catplot(x="day", y="total_bill", kind="boxen", data=df_exemplo);
sns.stripplot(x="day", y="total_bill", data=df_exemplo, color="blue");
```

## Plotar graficos em conjunto usando facetgrid, mostrando mais de um grafico em uma unica figura

```python
%matplotlib inline
sns.catplot(x="sex", y="total_bill", hue="smoker", col="time", data=df_exemplo, kind="boxen", height=4)
```
