# Usando técnicas de ML para analisar os dados do enem.

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%precision %.2f
pd.options.display.float_format = '{:,.2f}'.format
```

```python
uri = 'https://github.com/guilhermesilveira/enem-2018/blob/master/MICRODADOS_ENEM_2018_SAMPLE_43278.csv?raw=true'
dados =  pd.read_csv(uri)
dados.head()
```

```python
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']
#dados_notas = dados[colunas_notas]
dados_notas = dados[colunas_notas].dropna()
dados_notas.columns = ['ciencias_naturais', 'ciencias_humanas', 'linguagem_codigo', 'matematica', 'redacao']
dados_notas.head()
```

```python
tenho = dados_notas[['ciencias_naturais', 'ciencias_humanas','matematica', 'redacao']]
x_4_notas = tenho
tenho.head()
```

```python
adivinhar = dados_notas['linguagem_codigo']
y_adivinhar = adivinhar
adivinhar.head()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

np.random.seed(43267)

x_treino, x_teste, y_treino, y_teste = train_test_split(x_4_notas, y_adivinhar, random_state = 326784)

print(x_treino.shape)
print(x_teste.shape)
print(y_treino.shape)
print(y_teste.shape)
```

```python
from sklearn.svm import LinearSVR

modelo = LinearSVR()
modelo.fit(x_treino, y_treino)
```

```python
predicoes_notas_linguagem_codigo = modelo.predict(x_teste)
predicoes_notas_linguagem_codigo[:5]
```

```python
y_teste[:5]
```

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values, y=predicoes_notas_linguagem_codigo)
```

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values,  y = y_teste.values - predicoes_notas_linguagem_codigo)
```

# Testando o SVR puro - OBS.: não possui linearidade

```python
from sklearn.svm import SVR

modelo = SVR()
modelo.fit(x_treino, y_treino)
```

```python
predicoes_notas_linguagem = modelo.predict(x_teste)
plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values, y=predicoes_notas_linguagem_codigo)
```

# Usando arvore de decisão

```python
from sklearn.tree import DecisionTreeRegressor

modelo = DecisionTreeRegressor()
modelo.fit(x_treino, y_treino)
```

```python
predicoes_notas_linguagem = modelo.predict(x_teste)
plt.figure(figsize=(9,9))
sns.scatterplot(x=y_teste.values, y=predicoes_notas_linguagem)
```

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x=x_teste['matematica'].values, y=predicoes_notas_linguagem)
sns.scatterplot(x=x_teste['matematica'].values, y=y_teste.values)
```
